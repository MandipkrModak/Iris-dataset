```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Rename columns for better usability
iris_df.rename(columns={
    'sepal length (cm)': 'sl',
    'sepal width (cm)': 'sw',
    'petal length (cm)': 'pl',
    'petal width (cm)': 'pw'
}, inplace=True)

# Display basic info and summary statistics
print("DataFrame Info:")
print(iris_df.info())
print("\nSummary Statistics:")
print(iris_df.describe())

# Check for missing values
print("\nMissing Values:")
print(iris_df.isnull().sum())

# Distribution of species
print("\nSpecies Distribution:")
print(iris_df['target'].value_counts())

# Plot KDE for sepal length by species
sns.FacetGrid(iris_df, col='target', col_wrap=3, height=4, aspect=1).map(sns.kdeplot, 'sl', shade=True)
plt.suptitle('KDE Plot of Sepal Length for Each Species', y=1.02)
plt.show()

# Plot histograms for sepal length by species
sns.FacetGrid(iris_df, col='target', col_wrap=3, height=4, aspect=1).map(sns.histplot, 'sl', kde=False, bins=15, color='skyblue')
plt.suptitle('Histogram of Sepal Length for Each Species', y=1.02)
plt.show()

# Prepare data for machine learning
X = iris_df.drop('target', axis=1)  # Features
y = iris_df['target']               # Target variable

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train the Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions and evaluate the model
y_pred = model.predict(X_test)
print(f"\nModel Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

```
